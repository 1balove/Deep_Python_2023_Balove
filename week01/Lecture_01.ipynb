## Что такое git и для чего он нужен?

![git_logo](https://upload.wikimedia.org/wikipedia/commons/thumb/e/e0/Git-logo.svg/383px-Git-logo.svg.png)

Git — распределённая система управления версиями (спасибо вики). То есть с помощью этой системы можно отслеживать состояние и изменение текстовых файлов (кода) в базе данных.  

### Создание/клонирование git репозитория

Для того, чтобы локально создать git репозиторий, нужно зайти в каталог (папку) с вашим проектом и инициилизировать гит:
1. Открываем терминал
2. Заходим в папку с проектом:

    Linux:
    ```
    cd /home/user_name/path_to_project
    ```

    macOS:
    ```
    cd /Users/user_name/path_to_project
    ```

    Windows:
    ```
    cd C:/Users/user_name/path_to_project
    ```
3. Инициилизуем **git**:
    ```
    git init
    ```

В каталоге появится новый подкаталог с именем `.git`, содержащие структуру git репозитория.

Для того, чтобы склонировать репозиторий с git хранилища (github/gitlab), нужно ввести команду `git clone $url_to_repo$`

Например:

`git clone https://github.com/Palladain/Deep_Python.git`



### Состояние файлов в git репозитории

У файлов, которые вы создаете/изменяете существует 3 состояния:
1. **modified** - измененный файл (исходный файл, который был в последствии изменен)
2. **staged** - индексированный файл (это тот файл, у которого есть свой индекс)
3. **commited** - зафиксированный файл (это тот файл, который учитывается в гит репозитории)

![work_copy_git](https://git-scm.com/book/en/v2/images/areas.png)


Файлы также могут быть **отслеживаемыми** и **неотслеживаемыми**.

**Отслеживаемый** файл - это тот файл, о котором знает Git

**Неотслеживаемый** файл - это тот файл, о котором **НЕ** знает Git

### Жизненный цикл состояний файлов

![cycle_of_git_files](https://git-scm.com/book/en/v2/images/lifecycle.png)

1. Только созданный файл сразу индексируется (**staged** состояние)
2. Если у вас есть неостлеживаемый файл и вы хотите сделать его отслеживаемым, то надо использовать: ```git add file_example```
<!-- 3. Чтобы сделать файл неотслеживаемым, то требуется сначала сделать файл `.gitignore`.  -->
3. Чтобы зафиксировать изменения файлов, вводится команда `git commit -m `, где после `-m` надо написать сообщение о коммите. Требуется добавлять комментарии, отражающие краткую суть коммита.
Например, был добавлен файл **README**, и требуется фиксация файла.
Вводим `git commit -m add readme`.
4. Чтобы отследить состояние файлов, вводим `git status` и получаем:
```
$ git status
On branch main
Your branch is up-to-date with 'origin/main'.
Changes to be committed:
  (use "git reset HEAD <file>..." to unstage)

    new file:   README
```




### Просмотр индексированных и неиндексированных изменений

Если требуется посмотреть более подробную информацию об изменении состояния файла, то на помощь приходит команда `git diff`.

Например, мы хотим посмотреть более подробную информацию о **новом** файле **README**:

```
git diff --staged
diff --git a/README b/README
new file mode 164412
index 0000000..04233a5
--- /dev/null
+++ b/README
@@ -0,0 +1 @@
+My Project
```

Если мы хотим посмотреть информацию об **измененном** файле:

```
git diff
diff --git a/example_file.py b/example_file.py
index 6eab713..42f51e 202344
--- a/example_file.py
+++ b/example_file.py
@@ -13,2 +13,6 @@ branch directly, things can get messy.
 Please include a nice description of your changes when you submit your PR;
 if we have to read the whole diff to figure out why you're contributing
 in the first place, you're less likely to get feedback and have your change
-merged in.
+merged in. Also, split your changes into comprehensive chunks if you patch is
+longer than a dozen lines.
```


### Игнорируемые файлы

При разработки практически всегда (**всегда**) требуется игнорировать некоторые файлы, например:
1. .idea (это системные файлы вашей IDE)
2. .egg-info (это системные файлы какой-либо скомпилированной библиотеки)
3. какой-нибудь файл, который вы не хотите / не должны отслеживать.

И для этого требуется создать файл ```.gitignore```, который будет содержать названия файлов или паттер названия файлов.

Заходим в терминал и создаем файл:
```
cat .gitignore
```
Например, если вы работаете с каким-нибудь сервисом, который позволяет создавать git репозитории, то обычно ```.gitignore``` можно сгенерировать.

Основной паттерн команд для ```.gitignore```:



```
# 1. Исключить все файлы с расширением .ipynb
*.ipynb

# 2. Если требуется отслеживать файл example.ipynb, который попадает под ограничение #1, то добавляем в .gitignore:
!example.ipynb

# 3. Исключить файл TODO в корневом каталоге, но не файл в subdir/TODO
/TODO

# 4. Игнорировать весь каталог __pycache__/
__pycache__/

# 5. Игнорировать файл doc/notes.txt, но не файл doc/server/arch.txt
doc/*.txt

# 6. Игнорировать все .txt файлы в каталоге weights/
weights/**/*.txt
```

### Удаляемые файлы

Часто бывает при работе, что требуется удалить лишние файлы.

Например, ранее нами добавленный файл `file_example.py` не несет в себе никакой полезной информации:



```
$ rm file_example.py
$ git status
On branch main
Your branch is up-to-date with 'origin/main'.
Changes not staged for commit:
  (use "git add/rm <file>..." to update what will be committed)
  (use "git checkout -- <file>..." to discard changes in working directory)

        deleted:    file_example.py

no changes added to commit (use "git add" and/or "git commit -a")
```
Мы удалили файл, но не перестали отслеживать. Следовательно требуется добавить удаление в индекс:


```
$ git rm file_example.py
$ rm 'file_example.py'
$ git status
On branch main
Your branch is up-to-date with 'origin/main'.
Changes to be committed:
  (use "git reset HEAD <file>..." to unstage)

    deleted:    file_example.py
```




### Операции отката


Например, мы сделали коммит и поняли, что забыли проиндексировать изменения в файле, который хотели добавить в коммит, то можно добавить к `git commit` флаг `--amend`:

```
$ git commit -m 'Initial commit'
$ git add forgotten_file
$ git commit --amend
```
В итоге получится единый коммит — второй коммит заменит результаты первого.




Если требуется отменить индексацию после `git add example_python.py`:

```
$ git add example_python.py
$ git status
On branch main
Changes to be committed:
  (use "git reset HEAD <file>..." to unstage)

    new file:    example_python.py
```
То нужно ввести команду `git reset HEAD` (это прям написано в сообщении git'a)


```
$ git reset HEAD example_python.py
Unstaged changes after reset:
A	example_python.py
$ git status
On branch main

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git checkout -- <file>..." to discard changes in working directory)

    new file:   example_python.py
```

Ветки (branches) в системе контроля версий Git используются для разделения и управления различными версиями вашего проекта. Они предоставляют механизм для создания изолированных линий разработки, что позволяет работать над разными функциональными или исправительными задачами одновременно без вмешательства в основную линию разработки. Вот некоторые из основных целей использования веток в Git:

1. Разработка новых функций

2. Исправление ошибок

3. Совместная работа

4. Эксперименты и тестирование

Предположим, что мы накосячили и запушили в пятницу код, который положил нам прод.

Для того, чтобы проверить, какие ветки существуют, надо ввести команду ```git branch```:

```
git branch
* main
* iss53
```
Мы можем удостовериться, что никаких веток, кроме главной ветки ```main``` и ветки ```iss53```, в которой решается задача под номером 53.

Для того, чтобы исправить свой косяк, требуется создать отдельную ветку с незамысловатым названием "hotfix".


```
git checkout -b hotfix
Switched to a new branch 'hotfix'
git add /path_to_file/needed_file.py
git commit -m 'fix needed_file.py'
```
![hotfix_create_pic](https://git-scm.com/book/en/v2/images/basic-branching-4.png)

```git checkout -b hotfix```:
1. ```git checkout branch_name``` - переключение на нужно ветку
2. Флаг ```-b``` позволяет сразу создать ветку и переключиться на нее

Исправили ошибку и следовательно нам надо влить изменения в основную ```main``` ветку.
Для этого нам нужно:
1. Переключиться на ветку ```main```.
2. Выполнить команду ```git merge```.

```
$ git checkout main
Switched to branch 'main'
$ git merge hotfix
Merge made by the 'recursive' strategy.
needed_file.py |    322 +
1 file changed, 322 insertion(+)
```

После слияния нужно удалить ветку ```hotfix```:
```
$ git branch -d hotfix
Deleted branch hotfix (56s513j).
```

![hotfix_create_pic](https://git-scm.com/book/en/v2/images/basic-branching-5.png)

После отличных выходных, приходим в понедельник опять на РАБоту и садимся за задачу №53.
После того, как сделали задачу, нам опять надо влить изменения в ```main```, но произошел конфликт слияния.
```
$ git checkout main
Switched to branch 'main'
$ git merge iss53
Auto-merging file_to_fix.py
CONFLICT (content): Merge conflict in file_to_fix.py
Automatic merge failed; fix conflicts and then commit the result.
```
Есть два варианта решения ```merge conflict```:
1. Исправить файл вручную, то есть объединить все изменения из ветки ```main``` и ```iss53```
2. Использовать инструмент слияния, например ```opendiff```.

И после всех изменений опять запускаем ```git add``` и все готово к объединению.

### Работа с удаленным репозиторием

С вероятностью 99.9% вся разработка будет вестись в сервисе представляющий систему управления репозиториями кода для Git (GitHub/GitLab/Gitee).

И все репозитории в этих сервисах являются удаленными.

Для добавления удаленного репозитория, требуется ввести команду ```git remote add <shortname> <url>```.

```
$ git remote add python_course https://github.com/Palladain/Deep_Python
$ git remote -v

python_course	https://github.com/Palladain/Deep_Python (fetch)
python_course	https://github.com/Palladain/Deep_Python (push)

```


Для того, чтобы получить все изменения из удаленного репозитория, существует две команды:
1. ```git fetch``` - команда забирает с удаленного репозитория все новые данные и файлы, которых у вас нет.
2. ```git pull``` - команда забирает с удаленного репозитория все новые данные и сливает их с текущей. То есть одновременно выполняет ```git fetch``` и ```git merge```

Для того, чтобы отправить все внесенные изменения кода, требуется выполнить команду ```git push <remote-name> <branch-name>```.
```
git push remote_branch_name local_branch_name
```

## Что такое среды разработки и для чего они нужны

Виртуальная среда (или виртуальное окружение) - это изолированное пространство, в котором можно установить различные версии языков программирования, библиотек и зависимостей. Они предоставляют средство для создания и управления изолированными окружениями для разработки и выполнения программного кода. Вот несколько основных причин, почему виртуальные среды полезны:

1. Изоляция зависимостей: Виртуальные среды позволяют изолировать зависимости разных проектов. Это означает, что вы можете иметь разные версии библиотек и пакетов для разных проектов без конфликтов.

2. Управление зависимостями: Вы можете легко управлять зависимостями в виртуальных средах. Это включает в себя установку, обновление и удаление библиотек, а также создание списков зависимостей (например, файлов requirements.txt в Python) для точной репродукции окружения на другой машине.

3. Изоляция проектов: Каждый проект может иметь свою собственную виртуальную среду, что позволяет разработчикам работать над разными проектами с разными зависимостями на одной машине.

4. Легкость развертывания и переноса: Виртуальные среды упрощают развертывание проекта на других машинах. Вы можете передать информацию о среде (например, через файлы зависимостей) и легко воссоздать ее на другой машине.

### Venv

```venv``` - это сокращение от virtual environment в Python.

Это модуль, встроенный в стандартную библиотеку Python, который предоставляет средства для создания и управления виртуальными средами.

Виртуальные среды, созданные с использованием venv, представляют собой изолированные пространства, в которых можно устанавливать и управлять зависимостями и библиотеками для конкретных Python-проектов.

Основные команды:

1. Создание виртуальной среды:
```
python -m venv my_env_name
```
2. Активация виртуальной среды:

    Linux / macOS:

    ```
    source my_env_name/bin/activate
    ```

    Windows:

    ```
    my_env_name/Scripts/activate
    ```

3. Деактивация виртуальной среды:
```
deactivate
```
4. Установка библиотек/пакетов:
```
pip install package_name
```
5. Создание списка зависимостей:
Для сохранения списка зависимостей вашего проекта можно использовать команду pip freeze и перенаправить вывод в файл requirements.txt.
```
pip freeze > requirements.txt
```
6. Установка зависимостей из файла:
Для установки всех зависимостей из файла requirements.txt в виртуальной среде, выполните следующую команду:
```
pip install -r requirements.txt
```
7. Вы можете удалить виртуальную среду, просто удалив соответствующую папку. Однако будьте осторожны, чтобы не случайно удалить важные данные.

### Conda

Conda - это платформонезависимый установщик и менеджер пакетов, используемый в основном для управления библиотеками и зависимостями в средах Python, а также для создания и управления виртуальными средами. Conda широко используется в анализе данных, научных вычислениях и разработке программного обеспечения.

Вот несколько ключевых аспектов Conda:

1. Управление пакетами: Conda позволяет легко устанавливать, обновлять и удалять пакеты и библиотеки. Это полезно для управления зависимостями в Python-проектах и обеспечения совместимости версий.

2. Виртуальные среды: С помощью Conda вы можете создавать изолированные виртуальные среды, где каждая среда может иметь свои собственные зависимости и версии библиотек. Это позволяет избежать конфликтов между зависимостями разных проектов.

3. Платформонезависимость: Conda обеспечивает платформонезависимость, что означает, что вы можете использовать его на разных операционных системах, включая Windows, macOS и Linux.

4. Интеграция с другими инструментами: Conda может интегрироваться с другими инструментами, такими как Jupyter Notebook и инструменты для научных вычислений и анализа данных, что делает его популярным выбором в области науки о данных и машинного обучения.

Conda упрощает управление средами и зависимостями в проектах, позволяя разработчикам создавать и развертывать проекты с минимальными сложностями.

Основные команды:
1. Создание виртуальной среды:
```
conda create --name conda_env
```
где ```conda_env``` может быть любым названием, которое вы захотите дать виртуальному окружению.

2. Активация виртуальной среды:
```
conda activate conda_env
```
3. Декативация виртуальной среды:
```
conda deactivate
```
4. Создание виртуальной среды с определенной версией Python:
```
conda create --name conda_env python=version_of_Python
```
Например вы хотите создать окружение с версией 3.9:
```
conda create --name conda_env python=python3.9
```
5. Установка пакетов:
```
conda install name_of_packet
```
Например:
```
conda install pytorch torchvision torchaudio pytorch-cuda=11.8 -c pytorch -c nvidia
```
6. Удаление пакетов:
```
conda remove name_of_packet
```
Например:
```
conda remove torchaudio
```
7. Создание списка зависимостей (requirements.txt):
```
conda list --export > requirements.txt
```
8. Установка зависимостей из файла:
```
conda create --name name_of_conda_env --file requirements.txt
```
9. Удаление виртуальной среды:
```
conda env remove --name name_of_conda_env
```

Естественно, это вызвано человеческим фактором (мало того, что деревья вырубают, так еще и распространились пчелы-убийцы). Но есть радость: они очень хорошо рамножаются в неволе (на момент конца XX века было всего 70 особей) и что более радостно, они представляют 90% генетического разнообразия, что позволяет их выращивать без угрозы бесплодия и прочих проблем.
